```python
import numpy as np
import matplotlib.pyplot as plt

#Persamaan garis pertama melewati titik (a_12,b_12) dan titik (c_12,d_12)
#Berikut adalah titik (a_12,b_12)
a_12 = 6
b_12 = -2
#Berikut adalah titik (c_12,d_12)
c_12 = 12
d_12 = 2
#Mencari nilai gradien
m1_12 = d_12-b_12/c_12-a_12
print("Nilai gradien persamaan garis pertama = {:.2f}".format(m1_12))
#Mencari persamaan garis pertama y1 = m1_12*(x-a_12)+b_12
if -m1_12*a_12+b_12>0:
    print("Persamaan garis pertama adalah y1={:.2f}x+{:.2f} ".format(m1_12,-m1_12*a_12+b_12))
else:
    print("Persamaan garis pertama adalah y1={:.2f}x{:.2f} ".format(m1_12,-m1_12*a_12+b_12))

#Persamaan garis kedua melewati titik (e_12,f_12) dan titik (g_12,h_12)
#Berikut adalah titik (e_12,f_12)
e_12 = 6
f_12 = -2
#Berikut adalah titik (g_12,h_12)
g_12 = 2
h_12 = -3
#Mencari nilai gradien
m2_12 = h_12-f_12/g_12-e_12
print("Nilai gradien persamaan garis kedua = {:.2f}".format(m2_12))
#Mencari persamaan garis kedua y2 = m2_12*(x-e_12)+f_12
if -m2_12*e_12+f_12>0:
    print("Persamaan garis kedua adalah y2={:.2f}x+{:.2f} ".format(m2_12,-m2_12*e_12+f_12))
else:
    print("Persamaan garis kedua adalah y2={:.2f}x{:.2f} ".format(m2_12,-m2_12*e_12+f_12))

#Membuktikan nilai gradien persamaan garis pertama dan kedua
if m1_12==m2_12:
    print("Nilai Gradien persamaan garis pertama dan kedua sama, maka memopunyai solusi banyak atau tidak mempunyai solusi")
else:
    #Menentukan titik potong kedua garis tp_x_12 dan  tp_y_12
    tp_x_12 = (m1_12*a_12-b_12-m2_12*e_12+f_12)/(m1_12-m2_12)
    tp_y_12 = m2_12*(tp_x_12-e_12)+f_12
    print("Titik potong kedua garis adalah ({:.2f},{:.2f})".format(tp_x_12,tp_y_12))
#Menentukan jangkauan nilai x    
x = np.arange(tp_x_12-5,tp_x_12+10,0.25)
#Menentukan ukuran dan resolusi figure
plt.figure(figsize=(5, 5), dpi=120)
if -m1_12*a_12+b_12>0:
    plt.plot(x, m1_12*(x-a_12)+b_12, color='brown',label= 'y={:.2f}x+{:.2f}'.format(m1_12,-m1_12*a_12+b_12))
else:
    plt.plot(x, m1_12*(x-a_12)+b_12, color='brown',label= 'y={:.2f}x{:.2f}'.format(m1_12,-m1_12*a_12+b_12))
if -m2_12*e_12+f_12>0:
   plt.plot(x, m2_12*(x-e_12)+f_12, color='red',label = 'y={:.2f}x+{:.2f}'.format(m2_12,-m2_12*e_12+f_12))
else:
    plt.plot(x, m2_12*(x-e_12)+f_12, color='red',label = 'y={:.2f}x{:.2f}'.format(m2_12,-m2_12*e_12+f_12))    

plt.plot(tp_x_12,tp_y_12,color='blue',marker='o') #Menampilkan titik potong pada kedua garis
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
