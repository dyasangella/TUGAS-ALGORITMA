```python
import numpy as np
import matplotlib.pyplot as plt

a_12 = float(input("Nilai x dari titik pertama dari garis pertama adalah "))
b_12 = float(input("Nilai y dari titik pertama dari garis pertama adalah "))
c_12 = float(input("Nilai x dari titik kedua dari garis pertama adalah "))
d_12 = float(input("Nilai y dari titik kedua dari garis pertama adalah "))

e_12 = float(input("Nilai x dari titik pertama dari garis kedua adalah "))
f_12 = float(input("Nilai y dari titik pertama dari garis kedua adalah "))
g_12 = float(input("Nilai x dari titik kedua dari garis kedua adalah "))
h_12 = float(input("Nilai y dari titik kedua dari garis kedua adalah "))

m_1_12 = (d_12 - b_12) / (c_12 - a_12)  # Gradien garis pertama
m_2_12 = (h_12 - f_12) / (g_12 - e_12)  # Gradien garis kedua

# Persamaan garis pertama adalah y=m_1_12(x-a_12)+b_12=m_1_12x-m_1_12a_12+b_12
# Persamaan garis kedua adalah y=m_2_12(x-e_12)+f_12=m_2_12x-m_2_12e_12+f_12

c_1_12 = -m_1_12 * a_12 + b_12  # Titik potong garis pertama terhadap sumbu y
c_2_12 = -m_2_12 * e_12 + f_12  # titik potong garis kedua terhadap sumbu y garis kedua

print("Persamaan garis pertama adalah y={:.2f}x+{:.2f}".format(m_1_12, c_1_12))
print("Persamaan garis kedua adalah y={:.2f}x+{:.2f}".format(m_2_12, c_2_12))

if m_1_12 == m_2_12:
    # print("Kedua garis sejajar sehingga kemungkinannya adalah tidak terdapat solusi atau terdapat tak berhingga banyaknya solusi")
    if c_1_12 == c_2_12:
        print("Kedua garis berhimpitan sehingga akan terdapat tak berhingga banyaknya solusi")
        x_12 = np.linspace(-5, 5, 100)
        plt.plot(x_12, m_1_12 * (x_12 - a_12) + b_12, color='pink')
        plt.plot(x_12, m_2_12 * (x_12 - e_12) + f_12, color='orange')
        plt.show()
    else:
        print("Kedua garis tidak mungkin berpotongan, sehingga tidak akan terdapat solusi.")
        x_12 = np.linspace(-5, 5, 100)
        plt.plot(x_12, m_1_12 * (x_12 - a_12) + b_12, color='pink')
        plt.plot(x_12, m_2_12 * (x_12 - e_12) + f_12, color='orange')
        plt.show()
else:
    print("Akan terdapat satu titik potong")
    tp_x_12 = (-m_2_12 * e_12 + f_12 + m_1_12 * a_12 - b_12) / (m_1_12 - m_2_12)
    tp_y_12 = m_1_12 * (tp_x_12 - a_12) + b_12
    print("Titik potong kedua garis adalah ({:.2f},{:.2f})".format(tp_x_12, tp_y_12))
    
    x_12 = np.linspace(tp_x_12 - 5, tp_x_12 + 5, 100)
    plt.plot(x_12, m_1_12 * (x_12 - a_12) + b_12, color='pink')
    plt.plot(x_12, m_2_12 * (x_12 - e_12) + f_12, color='orange')
    plt.axvline(x=0, color='grey')
    plt.axhline(y=0, color='grey')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Garis melewati 2 titik
# 2. Garis diketahui gradien dan titik yang dilewati
# 3. Megikuti persamaan garis ax+by=c dan dx+ey=f

# Akan dicari titik potong dari dua garis yg memenuhi salah satu dari ketiga kasus di atas

kasus_12 = float(input("Kasus berapakah yang akan dikerjakan? "))

if kasus_12 == 1:
    print("Masing-masing garis akan melewati dua titik yang diketahui.")
    a_12 = float(input("Nilai x dari titik pertama dari garis pertama adalah "))
    b_12 = float(input("Nilai y dari titik pertama dari garis pertama adalah "))
    c_12 = float(input("Nilai x dari titik kedua dari garis pertama adalah "))
    d_12 = float(input("Nilai y dari titik kedua dari garis pertama adalah "))

    e_12 = float(input("Nilai x dari titik pertama dari garis kedua adalah "))
    f_12 = float(input("Nilai y dari titik pertama dari garis kedua adalah "))
    g_12 = float(input("Nilai x dari titik kedua dari garis kedua adalah "))
    h_12 = float(input("Nilai y dari titik kedua dari garis kedua adalah "))

    m_1_12 = (d_12 - b_12) / (c_12 - a_12)  # Gradien garis pertama
    m_2_12 = (h_12 - f_12) / (g_12 - e_12)  # Gradien garis kedua

    c_1_12 = -m_1_12 * a_12 + b_12  # Titik potong garis pertama terhadap sumbu y
    c_2_12 = -m_2_12 * e_12 + f_12  # Titik potong garis kedua terhadap sumbu y garis kedua

    print("Persamaan garis pertama adalah y={:.2f}x+{:.2f}".format(m_1_12, c_1_12))
    print("Persamaan garis kedua adalah y={:.2f}x+{:.2f}".format(m_2_12, c_2_12))

    if m_1_12 == m_2_12:
        if c_1_12 == c_2_12:
            print("Kedua garis berhimpitan sehingga akan terdapat tak berhingga banyaknya solusi")
            x_12 = np.linspace(-5, 5, 100)
            plt.plot(x_12, m_1_12 * (x_12 - a_12) + b_12, color='purple')
            plt.plot(x_12, m_2_12 * (x_12 - e_12) + f_12, color='orange')
            plt.axvline(x=0, color='grey')
            plt.axhline(y=0, color='grey')
            plt.show()
        else:
            print("Kedua garis tidak mungkin berpotongan, sehingga tidak akan terdapat solusi.")
            x_12 = np.linspace(-5, 5, 100)
            plt.plot(x_12, m_1_12 * (x_12 - a_12) + b_12, color='purple')
            plt.plot(x_12, m_2_12 * (x_12 - e_12) + f_12, color='orange')
            plt.axvline(x=0, color='grey')
            plt.axhline(y=0, color='grey')
            plt.show()
    else:
        print("Akan terdapat satu titik potong")
        tp_x_12 = (-m_2_12 * e_12 + f_12 + m_1_12 * a_12 - b_12) / (m_1_12 - m_2_12)
        tp_y_12 = m_1_12 * (tp_x_12 - a_12) + b_12
        print("Titik potong kedua garis adalah ({:.2f},{:.2f})".format(tp_x_12, tp_y_12))
    
        x_12 = np.linspace(tp_x_12 - 5, tp_x_12 + 5, 100)
        plt.plot(x_12, m_1_12 * (x_12 - a_12) + b_12, color='purple', label="y={:.2f}(x-{:.2f})+{:.2f}".format(m_1_12, a_12, b_12))
        plt.plot(x_12, m_2_12 * (x_12 - e_12) + f_12, color='orange', label="y={:.2f}(x-{:.2f})+{:.2f}".format(m_2_12, e_12, f_12))
        plt.scatter([tp_x_12], [tp_y_12], color='green', zorder=5)
        plt.text(tp_x_12, tp_y_12, '({:.2f}, {:.2f})'.format(tp_x_12, tp_y_12), fontsize=12, ha='right')
        plt.axvline(x=0, color='grey')
        plt.axhline(y=0, color='grey')
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
        
elif kasus_12 == 2:
    print("Masing-masing garis diketahui gradien dan titik yang dilewati.")
    m_1_12 = float(input("Gradien dari garis pertama adalah "))
    a_12 = float(input("Nilai x dari titik yang dilewati oleh garis pertama adalah "))
    b_12 = float(input("Nilai y dari titik yang dilewati oleh garis pertama adalah "))
    print("Persamaan garis pertama adalah y={:.2f}*(x-{:.2f})+{:.2f}".format(m_1_12, a_12, b_12))
    
    m_2_12 = float(input("Gradien dari garis kedua adalah "))
    c_12 = float(input("Nilai x dari titik yang dilewati oleh garis kedua adalah "))
    d_12 = float(input("Nilai y dari titik yang dilewati oleh garis kedua adalah "))
    print("Persamaan garis kedua adalah y={:.2f}*(x-{:.2f})+{:.2f}".format(m_2_12, c_12, d_12))
    
    c_1_12 = -m_1_12 * a_12 + b_12
    c_2_12 = -m_2_12 * c_12 + d_12
    
    if m_1_12 == m_2_12:
        print("Kedua garis sejajar")
        if c_1_12 == c_2_12:
            print("Terdapat tak berhingga banyaknya solusi")
            x_12 = np.linspace(-5, 5, 100)
            plt.plot(x_12, m_1_12 * (x_12 - a_12) + b_12, color='purple', label='Garis 1 & 2 (Berhimpitan)')
            plt.axvline(x=0, color='grey', linestyle='--', linewidth=0.5)
            plt.axhline(y=0, color='grey', linestyle='--', linewidth=0.5)
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("Tidak terdapat solusi")
            x_12 = np.linspace(-5, 5, 100)
            plt.plot(x_12, m_1_12 * (x_12 - a_12) + b_12, color='purple', label='Garis 1')
            plt.plot(x_12, m_2_12 * (x_12 - c_12) + d_12, color='orange', label='Garis 2')
            plt.axvline(x=0, color='grey', linestyle='--', linewidth=0.5)
            plt.axhline(y=0, color='grey', linestyle='--', linewidth=0.5)
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("Kedua garis akan berpotongan di satu titik.")
        tp_x_12 = (c_2_12 - c_1_12) / (m_1_12 - m_2_12)
        tp_y_12 = m_1_12 * (tp_x_12 - a_12) + b_12
        print("Titik potong kedua garis adalah ({:.2f},{:.2f})".format(tp_x_12, tp_y_12))
        
        x_12 = np.linspace(tp_x_12 - 5, tp_x_12 + 5, 100)
        plt.plot(x_12, m_1_12 * (x_12 - a_12) + b_12, color='purple', label="Garis 1: y={:.2f}(x-{:.2f})+{:.2f}".format(m_1_12, a_12, b_12))
        plt.plot(x_12, m_2_12 * (x_12 - c_12) + d_12, color='orange', label="Garis 2: y={:.2f}(x-{:.2f})+{:.2f}".format(m_2_12, c_12, d_12))
        plt.scatter([tp_x_12], [tp_y_12], color='green', zorder=5)
        plt.text(tp_x_12, tp_y_12, '({:.2f}, {:.2f})'.format(tp_x_12, tp_y_12), fontsize=12, ha='right')
        plt.axvline(x=0, color='grey', linestyle='--', linewidth=0.5)
        plt.axhline(y=0, color='grey', linestyle='--', linewidth=0.5)
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()

elif kasus_12 == 3:
    print("Kedua garis memenuhi persamaan berbentuk ax+by=c dan dx+ey=f.")
    a_12 = float(input("Koefisien dari x pada garis pertama adalah "))
    b_12 = float(input("Koefisien dari y pada garis pertama adalah "))
    c_12 = float(input("Konstanta dari garis pertama adalah "))
    
    d_12 = float(input("Koefisien dari x pada garis kedua adalah "))
    e_12 = float(input("Koefisien dari y pada garis kedua adalah "))
    f_12 = float(input("Konstanta dari garis kedua adalah "))
    
    det_12 = a_12 * e_12 - b_12 * d_12
    
    if det_12 == 0:
        print("Kedua garis akan sejajar atau berhimpitan")
        if (a_12 * f_12 - c_12 * d_12) == 0 and (b_12 * f_12 - c_12 * e_12) == 0:
            print("Kedua garis berhimpitan")
            x_12 = np.linspace(-5, 5, 100)
            y_12 = (c_12 - a_12 * x_12) / b_12
            plt.plot(x_12, y_12, color='purple', label='Garis berhimpitan')
            plt.axvline(x=0, color='grey', linestyle='--', linewidth=0.5)
            plt.axhline(y=0, color='grey', linestyle='--', linewidth=0.5)
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("Kedua garis sejajar")
            x_12 = np.linspace(-5, 5, 100)
            y1_12 = (c_12 - a_12 * x_12) / b_12
            y2_12 = (f_12 - d_12 * x_12) / e_12
            plt.plot(x_12, y1_12, color='purple', label='Garis 1')
            plt.plot(x_12, y2_12, color='orange', label='Garis 2')
            plt.axvline(x=0, color='grey', linestyle='--', linewidth=0.5)
            plt.axhline(y=0, color='grey', linestyle='--', linewidth=0.5)
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("Kedua garis akan berpotongan di satu titik.")
        tp_x_12 = (c_12 * e_12 - b_12 * f_12) / det_12
        tp_y_12 = (a_12 * f_12 - c_12 * d_12) / det_12
        print("Titik potong kedua garis adalah ({:.2f},{:.2f})".format(tp_x_12, tp_y_12))
        
        x_12 = np.linspace(tp_x_12 - 5, tp_x_12 + 5, 100)
        y1_12 = (c_12 - a_12 * x_12) / b_12
        y2_12 = (f_12 - d_12 * x_12) / e_12
        plt.plot(x_12, y1_12, color='purple', label='Garis 1')
        plt.plot(x_12, y2_12, color='orange', label='Garis 2')
        plt.scatter([tp_x_12], [tp_y_12], color='green', zorder=5)
        plt.text(tp_x_12, tp_y_12, '({:.2f}, {:.2f})'.format(tp_x_12, tp_y_12), fontsize=12, ha='right')
        plt.axvline(x=0, color='grey', linestyle='--', linewidth=0.5)
        plt.axhline(y=0, color='grey', linestyle='--', linewidth=0.5)
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
else:
    print("Hanya terdapat 3 kasus. Mohon masukkan 1, 2 atau 3 saja.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Koefisien untuk persamaan pertama: a_12*x + b_12*y = c_12
a_12 = 1
b_12 = 2
c_12 = 5

# Koefisien untuk persamaan kedua: d_12*x + e_12*y = f_12
d_12 = 2
e_12 = 3
f_12 = 6

# Periksa apakah matriks koefisien memiliki invers (determinannya tidak nol)
if a_12 * e_12 - b_12 * d_12 != 0:
    # Buat matriks koefisien dan matriks konstanta
    A_12 = np.array([[a_12, b_12], [d_12, e_12]])
    C_12 = np.array([[c_12], [f_12]])
    
    # Hitung invers dari matriks koefisien
    Ainv_12 = np.linalg.inv(A_12)
    print("Invers matriks A_12:")
    print(Ainv_12)
    
    # Hitung titik potong
    tp_xy_12 = np.matmul(Ainv_12, C_12)
    print("Titik potong:")
    print(tp_xy_12)
    
    # Visualisasi garis dan titik potong
    x_12 = np.linspace(-10, 10, 400)
    y1_12 = (c_12 - a_12 * x_12) / b_12
    y2_12 = (f_12 - d_12 * x_12) / e_12
    
    plt.plot(x_12, y1_12, label='Garis 1: {:.2f}x + {:.2f}y = {:.2f}'.format(a_12, b_12, c_12), color='orange')
    plt.plot(x_12, y2_12, label='Garis 2: {:.2f}x + {:.2f}y = {:.2f}'.format(d_12, e_12, f_12), color='blue')
    plt.scatter(tp_xy_12[0], tp_xy_12[1], color='green', zorder=5)
    plt.text(tp_xy_12[0], tp_xy_12[1], '({:.2f}, {:.2f})'.format(tp_xy_12[0,0], tp_xy_12[1,0]), fontsize=12, ha='right')
    plt.axvline(x=0, color='grey', linestyle='--', linewidth=0.5)
    plt.axhline(y=0, color='grey', linestyle='--', linewidth=0.5)
    plt.legend(loc='best')
    plt.grid(True)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Visualisasi Garis dan Titik Potong')
    plt.show()
else:
    print("Matrix singular, tidak memiliki invers dan tidak ada titik potong unik.")
```
