```python
import cmath
import numpy as np
import matplotlib.pyplot as plt

# Meminta input dari pengguna
akar1_12 = float(input("Masukkan akar pertama X1 = "))
akar2_12 = float(input("Masukkan akar kedua X2 = "))

while True:
    a_12 = float(input("Masukkan nilai a = "))
    if a_12 != 0:
        break
    else:
        print("Nilai a tidak boleh 0. Silahkan masukkan nilai a yang berbeda dari 0.")

# Menentukan persamaan kuadrat dari akar-akarnya
b_12 = -(akar1_12 + akar2_12)
c_12 = (akar1_12 * akar2_12)
print(f"Persamaan kuadratnya {a_12}x^2 + ({b_12}x) + ({c_12}) = 0")

# Menentukan titik puncak dan sumbu simetri
x_puncak_12 = -b_12.real / (2 * a_12)
y_puncak_12 = a_12 * x_puncak_12**2 + b_12 * x_puncak_12 + c_12
print(f"Titik puncaknya adalah {x_puncak_12},{y_puncak_12}")
print(f"Sumbu simetrinya adalah: x = {x_puncak_12}")

# Menentukan posisi titik puncak
if x_puncak_12 < 0:
    print("Titik puncak berada di kiri sumbu y.")
elif x_puncak_12 == 0:
    print("Titik puncak berada pada sumbu y.")
else:
    print("Titik puncak berada di kanan sumbu y.")

# Kondisi parabola memotong sumbu y
y_sumbu_y_12 = c_12
print(f"Parabola memotong sumbu y di titik: (0, {y_sumbu_y_12})")
if y_sumbu_y_12 > 0:
    print("Parabola memotong sumbu y positif")
elif y_sumbu_y_12 == 0:
    print("Parabola memotong sumbu y di pusat (0, 0)")
else:
    print("Parabola memotong sumbu y negatif")

# Macam-macam akar persamaan kuadrat dilihat dari diskriminan
D_12 = b_12**2 - ((4 * a_12) * c_12)
print(f"Nilai Diskriminan yaitu D = {D_12}")
if D_12.real > 0:
    print("Persamaan kuadrat memiliki dua akar real yang berbeda")
elif D_12.real == 0:
    print("Persamaan kuadrat memiliki dua akar real yang sama atau akar kembar")
else:
    print("Persamaan kuadrat memiliki dua akar tidak real")

# Penyelesaian persamaan kuadrat menggunakan rumus ABC
akar1_abc_12 = (-b_12 + cmath.sqrt(D_12)) / (2 * a_12)
akar2_abc_12 = (-b_12 - cmath.sqrt(D_12)) / (2 * a_12)
print(f"Akar-akarnya adalah {akar1_abc_12} dan {akar2_abc_12}")

# Menyusun persamaan kuadrat baru jika diketahui akar-akarnya
b_baru = -(akar1_abc_12 + akar2_abc_12).real
c_baru = (akar1_abc_12 * akar2_abc_12).real
print(f"Persamaan kuadrat baru dari akar-akarnya adalah {a_12}x^2 + ({b_baru})x + ({c_baru}) = 0")

# Menentukan grafik fungsi kuadrat
x_12 = np.linspace(-10, 10, 400)
y_12 = a_12 * x_12**2 + b_12 * x_12 + c_12

plt.figure(figsize=(8, 6))
plt.plot(x_12, y_12, label=f'y={a_12}x^2 + ({b_12})x + ({c_12})', color='orange')
plt.scatter(x_puncak_12, y_puncak_12, color='purple', marker='o', label=f"Titik Puncak = ({x_puncak_12}, {y_puncak_12})")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('Grafik Fungsi Kuadrat')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
